# Visualisation 

## 1. Closing Price
The closing price is the last price at which the stock is traded during the regular trading day. A stockâ€™s closing price is the standard benchmark used by investors to track its performance over time.

```python
plt.figure(figsize=(16, 8))
plt.title('Close Price History')
plt.plot(stock_cleaned_data.index, stock_cleaned_data['close'], linewidth=3, label='Close Price')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(loc='best')  # Display a legend if you have multiple lines on the chart
plt.grid(True)  # Add grid lines for reference
plt.show()
```

```python
sns.histplot(x='volume', data=stock_cleaned_data)
```

```python
corr_coeff = stock_cleaned_data['volume'].corr(stock_cleaned_data['close'])
print('Correlation Coefficient:', corr_coeff)
sns.jointplot(x='close', y='volume', data=stock_cleaned_data, kind='reg', 
              line_kws={'color': 'purple'})

plt.xlabel('Adj Close Price')
plt.ylabel('Volume')
```

```python
corr_coeff = stock_cleaned_data['volume'].corr(stock_cleaned_data['high'] - stock_cleaned_data['close'])
print('Correlation Coefficient:', corr_coeff)

sns.jointplot(x=stock_cleaned_data['high'] - stock_cleaned_data['close'], y='volume', data=stock_cleaned_data, kind='reg', 
              line_kws={'color': 'purple'})

plt.xlabel('high-low')
plt.ylabel('Volume')
```
